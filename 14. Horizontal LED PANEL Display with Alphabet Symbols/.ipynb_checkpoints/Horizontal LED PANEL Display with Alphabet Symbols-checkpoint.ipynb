```python
def print_row(*col,ln=False):
    for i in range(5):
        if i in col:
            print('*' , end =  ' ')
        else:
            print(' ' , end = ' ')
    
    if ln:
        print()
```

```python
sent = input("Enter any sentence : ").upper().split()
print()

for word in sent:
    for i in range(7):
        for char in word:
            eval(char+str(i)+'()'
            print(end=' ')
        print()
    print()
```
